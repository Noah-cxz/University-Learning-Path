### 分类性能指标概述
- 混淆矩阵：每一种预测值和真实值的组合
- 精确率和召回率：分别反映模型本身的命中性和在真实数据中的泛化能力
- F_1：综合精确率和召回率的形式化表达
- AUC面积：将正预测为正和将负预测为正的得分关系 

### 简单数据分析的流程和经验：
- 快速浏览数据集各个特征的范围
- 查看数据中**空值**的情况
- 查看**分类属性**以及该分类下的占比是否出现倾斜
- 一般的，根据属性的离散型和连续性划分为预测属性和分类属性。接下来分别探索这两大类特征
    - 数值分析：特征值占比、与目标变量的相关性
    - 可视化：单个特征的密度估计、多个特征组合的密度估计

# 特征工程

### 构造新特征，计算统计量，数据分桶

不确定将训练集和测试集合并构造特征的方式是否正确。在《机器学习实战》中处理标准化数据时，是不能够将test集的数据一并统计的

```python
car['train']=1
car_test['train']=0
data = pd.concat([car,car_test],ignore_index=True)
data.shape
```

```python
# 以day为单位计算使用时间
# 数据含有时间出错格式。 errors='coerce'将无法转换的格式设置为空值，随后需要检查一下空值的情况
data['used_time'] = (pd.to_datetime(data['creatDate'], format='%Y%m%d', errors='coerce') - 
                            pd.to_datetime(data['regDate'], format='%Y%m%d', errors='coerce')).dt.days
```

```python
# 不建议删除，因为删除缺失数据占总样本量过大：7.5%;另一方面，如果我们 XGBoost 之类的决策树，其本身就能处理缺失值
data['used_time'].isnull().sum()
```

```python
# 一共有四十种品牌
data['brand'].nunique()
```

```python
# 这里要以 train 的数据计算统计量(不加入test集的原因下面注释有解释)
car_gb = car.groupby("brand") # 使用分组
all_info = {}
for kind, kind_data in car_gb:
    info = {} 
#     print(kind) # 就是0-39的index
#     print(kind_data) # 每一个分类下的df的子集(样式还是df)
    # 写入字典：
    kind_data = kind_data[kind_data['price'] > 0]
    info['brand_amount'] = len(kind_data) # 每个分类的数量单独作为一个特征
    # 计算预测值price的统计量，所以不能加入test集
    # 常用的5~6个统计量
    info['brand_price_max'] = kind_data.price.max()
    info['brand_price_median'] = kind_data.price.median()
    info['brand_price_min'] = kind_data.price.min()
    info['brand_price_sum'] = kind_data.price.sum()
    info['brand_price_std'] = kind_data.price.std()
    info['brand_price_average'] = round(kind_data.price.sum() / (len(kind_data) + 1), 2)
    all_info[kind] = info
brand_fe = pd.DataFrame(all_info).T.reset_index().rename(columns={"index": "brand"})
data = data.merge(brand_fe, how='left', on='brand') # 按brand合并(brand)消失
data.head()
```

```python
# test集的统计量已经直接平滑过去
data['brand_price_min']
```

```python
# 数据分桶一般是为了增强模型的泛化性
# power这一属性的异常值是比较明显的
bin = [i*10 for i in range(31)] # 可以尝试其他组距、范围的数据分桶，观察那种参数下数据的分布更加均匀
data['power_bin'] = pd.cut(data['power'], bins=bin, labels=False)
data['power_bin'].plot.hist()
```

```python
# 删除冗余特征
data.drop(['creatDate','regDate'],inplace=True,axis=1)
```

```python
# 目前的数据其实已经可以给树模型使用
data.to_csv('../Used-Car-Price-Predict/data/data_for_tree.csv', index=0)
```

### 归一化和独热编码

```python
# power分布偏态，我们先取对数
data['power'] = np.log(data['power'] + 1) 
data['power'] = ((data['power'] - np.min(data['power'])) / (np.max(data['power']) - np.min(data['power'])))
data['power'].plot.hist()
```

```python
# 同样做标准化处理
data['kilometer'] = ((data['kilometer'] - np.min(data['kilometer'])) / 
                        (np.max(data['kilometer']) - np.min(data['kilometer'])))
data['brand_amount'] = ((data['brand_amount'] - np.min(data['brand_amount'])) / 
                        (np.max(data['brand_amount']) - np.min(data['brand_amount'])))
data['brand_price_average'] = ((data['brand_price_average'] - np.min(data['brand_price_average'])) / 
                               (np.max(data['brand_price_average']) - np.min(data['brand_price_average'])))
data['brand_price_max'] = ((data['brand_price_max'] - np.min(data['brand_price_max'])) / 
                           (np.max(data['brand_price_max']) - np.min(data['brand_price_max'])))
data['brand_price_median'] = ((data['brand_price_median'] - np.min(data['brand_price_median'])) /
                              (np.max(data['brand_price_median']) - np.min(data['brand_price_median'])))
data['brand_price_min'] = ((data['brand_price_min'] - np.min(data['brand_price_min'])) / 
                           (np.max(data['brand_price_min']) - np.min(data['brand_price_min'])))
data['brand_price_std'] = ((data['brand_price_std'] - np.min(data['brand_price_std'])) / 
                           (np.max(data['brand_price_std']) - np.min(data['brand_price_std'])))
data['brand_price_sum'] = ((data['brand_price_sum'] - np.min(data['brand_price_sum'])) / 
                           (np.max(data['brand_price_sum']) - np.min(data['brand_price_sum'])))
```

```python
data.shape
```

```python
# 对类别特征进行 OneEncoder
data = pd.get_dummies(data, columns=['model', 'brand', 'bodyType', 'fuelType',
                                     'gearbox', 'notRepairedDamage', 'power_bin'])# 使用独热编码，特征量虽然增大，但是泛化性会增强很多(实际测试不然)
```

```python
# 以下代码由Baseline提供，减少数据在内存中占用的空间
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() 
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() 
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    return df
car = reduce_mem_usage(pd.read_csv('data/data_for_tree.csv'))
```
